```python
H_sim = pd.read_excel(r'C:\\Users\\aawasthi\\OneDrive - Capgemini\\Desktop\\Work_Department\\Uniliver-Plants receipe\\Output\\Output_File.xlsx', sheet_name='Similarity')
H_sim_0 = H_sim.drop(['Grand Total'], axis=1)
n = 1 
H_sim_0 = H_sim_0.iloc[:-n]
H_sim_n = H_sim_0.set_index('Unnamed: 0')
H_sim_n = H_sim_n.astype(float)
H_sim_1 = H_sim_n[(H_sim_n[H_sim_n.columns] > 0.98) & (H_sim_n[H_sim_n.columns]< 1)]
sim_col = H_sim_1.idxmax(axis="columns")
sim_cols = pd.DataFrame(sim_col)
sim_column = sim_cols.reset_index(inplace=True)
H_sim_2 = pd.read_excel(r'C:\\Users\\aawasthi\\OneDrive - Capgemini\\Desktop\\Work_Department\\Uniliver-Plants receipe\\Output\\Output_File.xlsx', sheet_name='Country_ANALYSIS')
H_sim_3 = H_sim_2.filter(items=['Row Labels'])
H_sim_3 = H_sim_3.rename(columns = {'Row Labels': 'Hierarchy Modified GMC Commodity'}, inplace = False)
H_sim_3 = H_sim_3.iloc[:-2]
H_sim_3 = H_sim_3.set_index('Hierarchy Modified GMC Commodity')
chunk = pd.read_excel(r'C:\\Users\\aawasthi\\OneDrive - Capgemini\\Desktop\\Work_Department\\Uniliver-Plants receipe\\Output\\chunk.xlsx')
File = pd.read_excel(r'C:\\Users\\aawasthi\\OneDrive - Capgemini\\Desktop\\Work_Department\\Uniliver-Plants receipe\\Output\\Output_File.xlsx', sheet_name='Country_OUTPUT')
    
def chunk_sim():
    i=0
    H_sim_4 = pd.DataFrame(index= H_sim_3.index, columns= sim_cols.iloc[i]) 
    col1 = File.filter(items=['CUC Spec','Hierarchy Modified GMC Commodity','Percent of Item Quantity'])
    val1 =  col1['CUC Spec']==H_sim_4.columns[0]
    val2 = col1[val1]
    val2 = val2.filter(items=['Hierarchy Modified GMC Commodity','Percent of Item Quantity'])
    res =  pd.merge(H_sim_4, val2, how = 'outer', on=['Hierarchy Modified GMC Commodity'])
    res[res.columns[1]].fillna(res["Percent of Item Quantity"], inplace=True)
    res1 =  res.drop(['Percent of Item Quantity'], axis=1)
    val3 =  col1['CUC Spec']==H_sim_4.columns[1]
    val4 = col1[val3]
    val4 = val4.filter(items=['Hierarchy Modified GMC Commodity','Percent of Item Quantity'])
    res2 =  pd.merge(res1, val4, how = 'outer', on=['Hierarchy Modified GMC Commodity'])
    res2[res2.columns[2]].fillna(res2["Percent of Item Quantity"], inplace=True)
    res2 = res2.drop(['Percent of Item Quantity'], axis=1)
    res3 = res2.fillna(0)
    
    def concat(a, b):
        s1 = str(a)
        s2 = str(b)
        s = s1 + s2
        c = float(s)
        return c
    a = res3.columns[1]
    b = res3.columns[2]
    c = math.trunc(a)
    d = math.trunc(b)
    
    Sim = res3.min(axis=1)
    res3[concat(c,d)] = Sim
    chunk[res3.columns[1]] = res3[res3.columns[1]]  
    chunk[res3.columns[2]] = res3[res3.columns[2]]  
    chunk[res3.columns[3]] = res3[res3.columns[3]]  
    i+=1
    chunk.to_excel('C:\\Users\\aawasthi\\OneDrive - Capgemini\\Desktop\\Work_Department\\Uniliver-Plants receipe\\Output\\chunk.xlsx')
    return chunk
    
chunk_sim()    
```
